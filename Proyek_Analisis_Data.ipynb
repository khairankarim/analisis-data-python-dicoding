## Visualization & Explanatory Analysis

### Pertanyaan 1:
At what times are our services most used? (Kapan Jasa kita dipakai pengguna paling banyak?)

Summing all transaction on "casual", "registered", and "cnt" colomns<br>and grouping it into hour_visualization variable

```python
hour_visualization = hour_datadf.groupby(by="hr").agg({
    "casual": "sum",
    "registered": "sum",
    "cnt": "sum"})

# hour_visualization
# hour_visualization.head(25)
```

Create a List of Hours in a day in Hours_in_a_day variable

```python
hours_in_a_day = [x for x in range(24)]

casual_cols = hour_visualization["casual"]
registered_cols = hour_visualization["registered"]
cnt_cols = hour_visualization["cnt"]
# print(hours_in_a_day)
```

Visualizing Casual Transaction each hour

```python
Hours_bar = plt.bar(x=hours_in_a_day, height=casual_cols)
plt.title("Casual Transaction each hour")
plt.xlabel("Hour in a day")
plt.ylabel("Casual Transaction")
plt.ticklabel_format(axis='x', style='plain')
plt.show(Hours_bar)
casual_max = casual_cols.max()
print(casual_max)
# casual_cols[casual_max].value_counts()
#print("The mode of Casual Transaction is in {}".format())
```

Visualizing Registered Transaction each hour

```python
Hours_bar2 = plt.bar(x=hours_in_a_day, height=registered_cols)
plt.title("Registered Transaction each hour")
plt.xlabel("Hour in a day")
plt.ylabel("Registered Transaction")
plt.show(Hours_bar2)
```

Visualizing All Counted Transaction each hour

```python
Hours_bar3 = plt.bar(x=hours_in_a_day, height=cnt_cols)
plt.title("All Counted Transaction each hour")
plt.xlabel("Hour in a day")
plt.ylabel("All Counted Transaction")
plt.show(Hours_bar3)
```
